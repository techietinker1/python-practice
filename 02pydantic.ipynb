# 1. Basic Model


```python
from pydantic import BaseModel

class Student(BaseModel):
    name: str
    age: int

student = Student(name="Rupm", age=45)

print(student)
print(student.name)
print(student.age)
```

# 2. Type Validation


```python
from pydantic import BaseModel

class User(BaseModel):
    name: str
    age: int

user = User(name="Rupm", age="45")

print(user)
```

# 3. Validation Error Example


```python
from pydantic import BaseModel

class Product(BaseModel):
    name: str
    price: float

product = Product(name="Laptop", price="abc")

print(product)
```

# 4. Optional Field


```python
from pydantic import BaseModel
from typing import Optional

class Employee(BaseModel):
    name: str
    salary: float
    department: Optional[str] = None

emp = Employee(
    name="Rupm",
    salary=50000
)

print(emp)
```

# 5. Default Value


```python
from pydantic import BaseModel

class Phone(BaseModel):
    brand: str
    price: int
    in_stock: bool = True

phone = Phone(
    brand="Samsung",
    price=20000
)

print(phone)
```

# 6. Email Validation


```python
from pydantic import BaseModel, EmailStr

class User(BaseModel):
    name: str
    email: EmailStr

user = User(
    name="Rupm",
    email="rupm@gmail.com"
)

print(user)
```

# 7. Field Constraints


```python
from pydantic import BaseModel, Field

class Student(BaseModel):
    name: str
    age: int = Field(gt=18, lt=30)

student = Student(
    name="Rupm",
    age=29
)

print(student)
```

# 8. List Example


```python
from pydantic import BaseModel

class Classroom(BaseModel):
    students: list[str]

room = Classroom(
    students=["Rupm", "Arpit", "Antra"]
)

print(room)
```

# 9. Nested Models


```python
from pydantic import BaseModel

class Address(BaseModel):
    city: str
    pin: int

class Student(BaseModel):
    name: str
    address: Address

student = Student(
    name="Rupm",
    address={
        "city": "Bermuda",
        "pin": 1111
    }
)

print(student)
```

# 10. Convert Model to Dictionary


```python
from pydantic import BaseModel

class User(BaseModel):
    name: str
    age: int

user = User(
    name="Rupm",
    age=45
)

print(user.model_dump())
```

# 11. JSON Output


```python
from pydantic import BaseModel

class User(BaseModel):
    name: str
    age: int

user = User(
    name="Rupm",
    age=45
)

print(user.model_dump_json())
```
